```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ipl = pd.read_csv('matches.csv')
ipl.head()
```

```python
ipl['player_of_match'].value_counts()
```

```python
#ONLY NAMES
list(ipl['player_of_match'].value_counts()[0:5].keys())
```

```python
plt.figure(figsize=(16,9))
plt.bar(list(ipl['player_of_match'].value_counts()[0:5].keys()),list(ipl['player_of_match'].value_counts()[0:5]),color='g')
plt.show()
```

```python
ipl['result'].value_counts()
```

```python
ipl['toss_winner'].value_counts()
```

```python
batting_first_wins = ipl[ipl['win_by_runs'] !=0]
batting_first_wins.head()
```

```python
plt.figure(figsize=(8,6))
plt.hist(batting_first_wins['win_by_runs'])
plt.title('batting first winner')
plt.xlabel('no. of runs')
plt.ylabel('no. of matches')
plt.show()
```

```python
batting_first_wins['winner'].value_counts()
```

```python
plt.figure(figsize=(16,9))
plt.bar(list(batting_first_wins['winner'].value_counts()[0:3].keys()),list(batting_first_wins['winner'].value_counts()[0:3]),color = ['b','g','orange'])

plt.show()
```

```python
plt.figure(figsize=(16,9))
plt.pie(list(batting_first_wins['winner'].value_counts()),labels=list(batting_first_wins['winner'].value_counts().keys()),autopct='%0.1f%%' 

plt.show()
```

```python
wine_reviews_points_mean = wine_reviews.points.mean()
def remean_points(row):
    row.points = row.points - wine_reviews_points_mean
    return row

wine_reviews.apply(remean_points, axis = 'columns')
```

```python
pp = wine_reviews.points/wine_reviews.price
a = pp.idxmax()
best_wine = wine_reviews.loc[a,'title']
print(best_wine)
```

```python
wine_reviews.description.str.count('tropical').sum()
wine_reviews.country.loc()
```

```python
def rating(row):
    return '***' if row.points >= 95 or row.country == 'Canada' else ('**' if row.points >= 85 else '*')

star_ratings = wine_reviews.apply(rating, axis='columns')
print(star_ratings)
        
        
```

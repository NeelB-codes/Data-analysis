```python
import pandas as pd
import matplotlib.pyplot as pl
import seaborn as sns
```

```python
fifa =pd.read_csv('players_20.csv')
fifa.head()
```

```python
for a in fifa.columns:
    print(a)
```

```python
fifa.shape
```

```python
fifa['nationality'].value_counts()[0:5]
```

```python
pl.figure(figsize=(7,8))
pl.bar(list(fifa['nationality'].value_counts()[0:5].keys()),list(fifa['nationality'].value_counts()[0:5]),color=['grey','black','red','blue','cyan'])
pl.show()
```

```python
wages=fifa[['short_name','wage_eur']]
wages.head()
```

```python
highest_paid_list = wages.sort_values(by=['wage_eur'],ascending=False)
highest_paid_list.head()
```

```python
pl.figure(figsize=(16,9))
pl.bar(list(highest_paid_list['short_name'])[0:5],list(highest_paid_list['wage_eur'])[0:5],color='g')
pl.show()
```

```python
fifa.nationality['Germany']
```

```python
German = fifa[fifa['nationality']=='Germany']
German.head()
```

```python
tallest_german_player = German.sort_values(by='height_cm',ascending=False)
tallest_german_player.head()
```

```python
heaviest_german_players = German.sort_values(by='weight_kg',ascending=False)
heaviest_german_players.head()
```

```python
highest_paid_german = German.sort_values(by='wage_eur',ascending=False)
highest_paid_german.head()
```

```python
german_wages = German[['short_name','wage_eur']].sort_values(by='wage_eur',ascending=False)
```

```python
pl.figure(figsize=(7,7))
pl.bar(list(german_wages['short_name'])[0:5],list(german_wages['wage_eur'])[0:5])
pl.show()
```

```python
best_shooters = fifa[['short_name','shooting']].sort_values(by='shooting',ascending=False)
best_shooters.head()
```
